```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F

import argparse
import os
import sys
from torchvision import datasets, transforms

import wandb
from tqdm import tqdm
    
```

```python
class FCN(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(FCN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, input_size)


    def forward(self, x):
        ##flatten x
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
argparser = argparse.ArgumentParser()
argparser.add_argument('-d','--input_size', type=int, default=28*28, help='model name')
argparser.add_argument('-n','--hidden_size', type=int, default=8192, help='hidden size')


argparser.add_argument('-r','--range', type=float, default=10.0, help='set the range of the input data')
argparser.add_argument('-N','--num-samples', type=int, default=32, help='set the number of the input data centers')
argparser.add_argument('-a','--alpha', type=float, default=90, help='detenrmine the angle between close input data points')
argparser.add_argument('-s','--sigma', type=float, default=0.05, help='noise level')

argparser.add_argument('-S','--seed', type=int, default=0, help='set the random seed')
argparser.add_argument('-D','--device', type=int, default=0, help='set the device')

argparser.add_argument('-x','--dist', type=int, default=1.0, help='set the distance between the two neighbouring input data points')

argparser.add_argument('--lr', type=float, default=0.1, help='learning rate')
argparser.add_argument('--wd', type=float, default=0.1, help='weight decay')
argparser.add_argument('-b','--batch-size', type=int, default=128, help='batch size')
argparser.add_argument('-E','--epochs', type=int, default=1000, help='epochs')

argparser.add_argument('--misc', action='store_true', default=False, help='miscellaneous')

argparser.add_argument('-o','--obtuse', action='store_true', default=False, help='obtuse angle in the triangle')


argparser.add_argument('--cosine', action='store_true', default=False, help='use cosine learning rate')

argparser.add_argument('-e','--predict-e', action='store_true', default=False, help='predict epsilon')


argparser.add_argument('--test-samples', type=int, default=1000, help='set the number of the test samples')
```

```python
def add_weight_decay(model, weight_decay=1e-5):
    decay = []
    no_decay = []
    for name, param in model.named_parameters():
        if not param.requires_grad:
            continue  # frozen weights
        if len(param.shape) == 1 or name.endswith(".bias"):
            no_decay.append(param)
        else:
            decay.append(param)
    return [
        {'params': no_decay, 'weight_decay': 0.},
        {'params': decay, 'weight_decay': weight_decay}]

def create_mnist_datapoints(classes=None, mindeg=0, maxdeg=180):
    
    transform=transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.1307,), (0.3081,))
        ])

    ds = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
    
    
    num_samples = len(ds)
    deg = -1
    while deg < mindeg or deg > maxdeg:
        if classes is None or len(classes) == 0:
            idx1, idx2, idx3 = np.random.choice(num_samples, 3, replace=False)
            x1, _  = ds[idx1]
            x2, _ = ds[idx2]
            x3, _ = ds[idx3]
        else:
            idx1 = np.random.randint(num_samples)
            x1, y1  = ds[idx1]
            while y1 not in classes:
                idx1 = np.random.randint(num_samples)
                x1, y1  = ds[idx1]
            
            idx2 = np.random.randint(num_samples)
            x2, y2  = ds[idx2]
            while y2 not in classes or idx2 == idx1:
                idx2 = np.random.randint(num_samples)
                x2, y2  = ds[idx2]
            
            idx3 = np.random.randint(num_samples)
            x3, y3  = ds[idx3]
            while y3 not in classes or idx3 == idx1 or idx3 == idx2:
                idx3 = np.random.randint(num_samples)
                x3, y3  = ds[idx3]


        d1 = x1-x2
        d2 = x1-x3
        d3 = x2-x3

        deg1 = torch.arccos(torch.sum(d1*d2)/torch.sqrt(torch.sum(d1*d1)*torch.sum(d2*d2)))*180/np.pi
        deg2 = torch.arccos(torch.sum(-d1*d3)/torch.sqrt(torch.sum(d1*d1)*torch.sum(d3*d3)))*180/np.pi
        deg3 = 180 - deg1 - deg2

        if deg3 >= deg1 and deg3 >= deg2:
            x0 = x3
            deg, v0, v1 = deg3, -d3, -d2
        elif deg2 >= deg1:
            x0 = x2
            deg, v0, v1 = deg2, -d1, d3
        else:
            x0 = x1
            deg, v0, v1 = deg1, d1, d2

    x1, x2 ,x3 = x1-x0, x2-x0, x3-x0
    x1 = x1.view(1, -1)
    x2 = x2.view(1, -1)
    x3 = x3.view(1, -1)
    v0 = v0.view(1, -1)
    v1 = v1.view(1, -1)
    X = torch.cat((x1,x2,x3), dim=0)

    print("degs = ", deg1, ", ", deg2, ", ", deg3)

    return X, deg, v0, v1


def deg2rad(deg):
    return deg*np.pi/180.0


def create_datapoints(num_samples, input_size, mrange, deg, dist, device):
    alpha = deg2rad(deg)

    datapoints = (torch.rand(num_samples, input_size, device=device)-0.5)*(mrange*2)

    V1 = torch.randn(num_samples, input_size, device=device)
    V2 = torch.randn(num_samples, input_size, device=device)
    V2 = V2 - torch.sum(V2*V1, dim=1, keepdim=True)*V1/torch.sum(V1*V1, dim=1, keepdim=True)
    V1 = V1/torch.sqrt(torch.sum(V1*V1, dim=1, keepdim=True))
    V2 = V2/torch.sqrt(torch.sum(V2*V2, dim=1, keepdim=True))

    datapoints_2 = V1*dist
    datapoints_3 = args.dist*(V1*np.cos(alpha) + V2*np.sin(alpha))

    datapoints = torch.cat((datapoints, datapoints_2, datapoints_3), dim=0)

    ##shuffle the data points:
    datapoints = datapoints[torch.randperm(datapoints.shape[0])]
    return datapoints
```

```python
txt = "--misc -E 30000 -s 0.5"
args = argparser.parse_args(txt.split())
```

```python
#set the random seed
torch.manual_seed(args.seed)

#set the device
device = torch.device("cuda:"+str(args.device) if torch.cuda.is_available() else "cpu")

#set the input data

if args.obtuse:
    X ,deg, v0, v1  = create_mnist_datapoints(classes = [0], mindeg=95, maxdeg=180)
else:
    X ,deg, v0, v1  = create_mnist_datapoints(classes = [0], mindeg=0, maxdeg=85)
    
print("degree: ", deg)

X = X.to(device)
v0 = v0.to(device)
v1 = v1.to(device)
```

```python
#set the model
model = FCN(args.input_size, args.hidden_size).to(device)

#set the optimizer
paramlist = add_weight_decay(model, weight_decay=1e-5)
optimizer = torch.optim.Adam(paramlist, lr=1E-5)


#set the loss function
loss_func = nn.MSELoss()

#set the wandb
if not args.misc:
    wandb.init(project="phased_denoiser", entity='dl-projects' ,  config=args)
    wandb.watch(model, log="all")


##create a data loader from the data points:

#set the training loop



repeats = (3+args.batch_size-1)//3

```

```python
bar = tqdm( range(args.epochs), total = args.epochs, desc="training")


for epoch in bar:
    model.train()
    total_loss = 0


    x = X.repeat(repeats, 1)
    x = x[torch.randperm(x.shape[0])[:args.batch_size]]

    optimizer.zero_grad()

    noise = torch.randn_like(x)
    y = x + args.sigma* noise


    if args.predict_e:
        eat = model(y)
        loss = loss_func(eat, noise )
    else:
        xat = model(y)
        loss = loss_func(xat, x)


    loss.backward()
    optimizer.step()


    total_loss = loss.item()


    if wandb.run is not None:
        wandb.log({"epoch": epoch, "loss": total_loss, 'learning rate': optimizer.param_groups[0]['lr']})

    bar.set_postfix(loss=total_loss)
```

```python


def project(x, u0, u1):
    x0 = torch.sum(x*u0, dim=1, keepdim=True)
    x1 = torch.sum(x*u1, dim=1, keepdim=True)
    x2 = x - x0*u0 - x1*u1
    x_orth = torch.sqrt(torch.sum(x2*x2, dim=1, keepdim=True))
    return x0, x1, x_orth

def plot_points_clean(X, u0, u1, v0 ,v1, ax, color = 'w'):
    x0, x1 , _ = project(X[0], u0, u1)
    ax.scatter(x0.cpu()/v0.norm().cpu().item(), x1.cpu()/v1.norm().cpu().item(), color=color)
    x0, x1 , _ = project(X[1], u0, u1)
    ax.scatter(x0.cpu()/v0.norm().cpu().item(), x1.cpu()/v1.norm().cpu().item(), color=color)
    x0, x1 , _ = project(X[2], u0, u1)
    ax.scatter(x0.cpu()/v0.norm().cpu().item(), x1.cpu()/v1.norm().cpu().item(), color=color)

def plot_points(X, u0, u1, v0 ,v1, ax,sigma, N = 128,  color = 'w'):
    x = X.repeat((3+N-1)//3, 1)
    x = x[torch.randperm(x.shape[0])[:N]]

    noise = torch.randn_like(x)
    y = x + sigma* noise

    for i in range(y.size(0)):
        x0, x1 , _ = project(y[i], u0, u1)
        ax.scatter(x0.cpu()/v0.norm().cpu().item(), x1.cpu()/v1.norm().cpu().item(), color=color, s=0.1 )

```

```python

grid_range = 3
grid_points = 100

gridvals = torch.linspace(-grid_range, grid_range, grid_points, device=device)

u0 = v0 / torch.sqrt(torch.sum(v0*v0, dim=1, keepdim=True))
u1 = v1 / torch.sqrt(torch.sum(v1*v1, dim=1, keepdim=True))


X0 = torch.zeros(grid_points, grid_points)
X1 = torch.zeros(grid_points, grid_points)
X_orth = torch.zeros(grid_points, grid_points)


for i, a1 in tqdm(enumerate(gridvals), total = grid_points):
    for j, a2 in enumerate(gridvals):
        y = a1*v0 + a2*v1
        
        xhat = model(y.view(1, -1))
        
        x0, x1 ,x_orth = project(xhat, u0, u1)
        
        X0[i,j]=x0.item()
        X1[i,j]=x1.item()
        X_orth[i,j]=x_orth.item()


Y = np.asarray(np.meshgrid(gridvals.cpu().numpy(), gridvals.cpu().numpy()))

X0 = X0.cpu().numpy()
X1 = X1.cpu().numpy()
X_orth = X_orth.cpu().numpy()
```

```python

levels = np.linspace(-100, 100, 21)
fig, ax = plt.subplots(3,  figsize=(8,12),sharex='col', sharey='row', layout='constrained')

CS = ax[0].contourf(Y[0], Y[1],X0, levels = levels , title = "Direction 0")
plot_points(X, u0, u1, v0, v1, ax[0], sigma = args.sigma)

# Add the contour line levels to the colorbar

# plt.figure()

ax[1].contourf(Y[0], Y[1],X1, levels = levels, title = "Direction 1")
plot_points(X, u0, u1, v0, v1, ax[1], sigma = args.sigma)

# plt.figure()


ax[2].contourf(Y[0], Y[1],X_orth, levels = levels, title = "Orthogonal Direction")
plot_points(X, u0, u1, v0, v1, ax[2], sigma = args.sigma)



# plt.figure()

plt.colorbar(CS, ax=ax)
plt.savefig(f"figures/contour_deg_{int(np.round(deg))}.pdf")
```
